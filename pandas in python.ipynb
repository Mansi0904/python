```python
data = {
    'PassengerId': [1, 2, 3, 4, 5],
    'Survived': [0, 1, 1, 1, 0],
    'Pclass': [3, 1, 3, 1, 3],
    'Name': ['Braund, Mr. Owen Harris', 'Cumings, Mrs. John Bradley', 'Heikkinen, Miss. Laina', 'Futrelle, Mrs. Jacques Heath', 'Allen, Mr. William Henry'],
    'Sex': ['male', 'female', 'female', 'female', 'male'],
    'Age': [22.0, 38.0, 26.0, 35.0, 35.0],
    'SibSp': [1, 1, 0, 1, 0],
    'Parch': [0, 0, 0, 0, 0],
    'Ticket': ['A/5 21171', 'PC 17599', 'STON/O2. 3101282', '113803', '373450'],
    'Fare': [7.2500, 71.2833, 7.9250, 53.1000, 8.0500],
    'Cabin': [None, 'C85', None, 'C123', None],
    'Embarked': ['S', 'C', 'S', 'S', 'S']
}

df = pd.DataFrame(data)
```

```python
df['Pclass'].value_counts().sort_index().plot(kind='bar', figsize=(8, 6))
plt.title('Passenger Class Distribution')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
df['Age'].plot(kind='hist', bins=20, figsize=(8, 6))
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Sex'].value_counts().plot(kind='bar', figsize=(8, 6))
plt.title('Passenger Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Bar Plot of Passenger Class
plt.figure(figsize=(8, 6))
df['Pclass'].value_counts().sort_index().plot(kind='bar')
plt.title('Passenger Class Distribution')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
# Histogram of Ages
plt.figure(figsize=(8, 6))
df['Age'].plot(kind='hist', bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Count of Passengers by Gender
plt.figure(figsize=(8, 6))
df['Sex'].value_counts().plot(kind='bar')
plt.title('Passenger Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Scatter Plot of Age vs Fare
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Fare'], alpha=0.7)
plt.title('Scatter Plot: Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
# Box Plot of Fare by Passenger Class
plt.figure(figsize=(8, 6))
df.boxplot(column='Fare', by='Pclass')
plt.title('Box Plot: Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.suptitle('')  # Remove default title
plt.show()
```

```python
# Pie Chart of Survival Rate
plt.figure(figsize=(8, 6))
survival_counts = df['Survived'].value_counts()
plt.pie(survival_counts, labels=['Not Survived', 'Survived'], autopct='%1.1f%%', startangle=90)
plt.title('Survival Rate')
plt.show()
```

```python
#observations
'''
1.Passenger Class Distribution (Bar Plot):

Most of the passengers in the dataset belong to the 3rd class (Pclass=3).
The number of passengers decreases from 1st class to 3rd class.
2.Age Distribution (Histogram):

The age distribution appears to be somewhat right-skewed.
The majority of passengers are between the ages of 20 and 40.
3.Passenger Gender Distribution (Bar Plot):

There are more male passengers than female passengers in the dataset.
Gender distribution is not equal, with a larger number of males.
4.Scatter Plot: Age vs Fare:

There is no strong linear correlation between age and fare.
Fares vary across different age groups, but there is no clear trend.
5.Box Plot: Fare by Passenger Class:

The box plot shows that the fares for 1st class tend to be higher compared to 2nd and 3rd classes.
There are potential outliers in each class, particularly in the 1st class.
6.Pie Chart: Survival Rate:

The pie chart illustrates the distribution of survival outcomes.
A higher percentage of passengers did not survive compared to those who survived.


'''
```
